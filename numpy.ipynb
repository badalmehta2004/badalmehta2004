```python
import numpy as np
```

```python
a1=np.array([1,2,3,4]) # 1 1 d array
a2=np.array([[1,2,3,4],[5,6,7,8]]) # 2d array list of multiple 1d array
a3=np.array([[[1,2],[3,4]],[[5,6],[7,8]]]) # 3d array multiple 2d array
print(a1)
print(a2)
print(a3)
```

```python
# default values
# np.zeros(shape)
a4=np.zeros(4) # 1 d
print(a4)
a5=np.zeros((2,3)) # 2d
print(a5)

# note if array toh homogenous data hi store kargea \
# if multiple datatypes denge toh jo superior datatype hoga vo hoga array a dataype
# np.ones(shape)

a4=np.ones(4,dtype='i4') #
print(a4)
a5=np.ones((2,3)) # 2d
a5
a4.dtype
```

```python
# specific value
#np.full(shape,value)
a6=np.full(4,10)
print(a6)

a7=np.full((3,4),5) # 3 X 4 matrix and each value is 5
print(a7)

a7=np.full((3,4,2),5) # 3 X 4 matrix and each value is 5
print(a7)
```

```python
# identity matrix
#np.eye(size) # alaways square matrix

idetity_matrix=np.eye(4)
print(idetity_matrix)
```

```python
# sequence of numbers
# arange(start,end,step) returns numpy array same as range() which return list
a9=np.arange(1,11)
print(a9)
a9=np.arange(0,11,2) # even number
print(a9)
```

```python
# properties of numpy array
# shape rows,cols
#array_name.attribute
print(a1.shape)
print(a2.shape)
print(a3.shape)

# size # number of elements

print(a1.size)
print(a2.size)
print(a3.size)

# ndim
# number of dimension
print(a1.ndim)
print(a2.ndim)
print(a3.ndim)
```

```python
# dtype

print(a1.dtype)
print(a2.dtype)
print(a3.dtype)
# a1.dtype='float32' # changed the datatype of a1 to float 32
# print(a1)

# change dtype
#.astype(new_dtype)
new_a1=a1.astype(float)
print(new_a1)
print(new_a1.dtype)
```

```python
# maths opertaions har ek element pe ho jayega
a10=np.array([10,20,30,40,50,60,70])
print(a10+5)
print(a10-5)
print(a10*2)
print(a10/2)
print(a10**3)
print(a10//4)
print(a10%3)
# print(a1+a10)
print(a1)
print(a10)
# print(a1*a10) # not matrix multiplcation its corresponding element multiplication
```

```python
# matrix mul
a11=np.array([[1,2],[3,4],[5,6]]) # 3 X 2
a12=np.array([[1,2,3],[4,5,6]])  # 2 X 3
print(a11@a12) # @ sign is used for matrix multiplication
print(np.matmul(a11,a12)) # matmul method
```

```python
# agreegation functions
print(a10)
print(np.sum(a10))
print(np.mean(a10))
print(np.min(a10))
print(np.max(a10))
print(np.std(a10))
print(np.var(a10))
```

```python
#indexing
print(a10)
print(a10[0]) # ist element
print(a10[-1]) # last element
print(a10[1]) # 2nd element
print(a10[-2]) # second last element
```

```python
# slicing
# array[start,end,step] step has direction and jump that is step by default right direction and step=1

# whole array
print(a10[0::])
# every 2nd element\
print(a10[0::2])
#revrese
print(a10[::-1])
# any middle values
print(a10[1:4])
```

```python
# fancy indexing
# selecting multiple indexes at once
print(a10[[1,4,6]]) # we give list of indexes in index field
# creates duplicate array of indexs given
```

```python
# booklean indexing
# condition on rows
print(a10[a10>30]) # all values greater than 30
print(a10[a10>25]) # all values greater than 25
```

```python
#reshaping and manipulating arrays
#reshaping change dimesion of array without modify data 1D ->2D 2D ->3D
# rehaping returns the view of original array  not copy changing in reshaped array also affects original array
print(a10)
reshaped_a10=a10.reshape(7,1)
print(reshaped_a10)
a13=np.array([1,2,3,4,5,6])
a13_reshaped=a13.reshape(3,2)
print(a13_reshaped)
a13_reshaped[1][1]=50 # changed in reshaped array
print(a13_reshaped)
print(a13.reshape(2,3))
print(a13) # 50 is also reflected in original array
```

```python
# flattening array
# multidimension to 1D

# 1 flatten array retuns copy chnaging value Does not affect original array
#.flatten()
# a13_reshaped_flat=a13_reshaped.flatten()
# print(a13_reshaped_flat)
# a13_reshaped_flat[1]=100 # changed in flatten array
# print(a13_reshaped_flat)
# print(a13) # does not affect original array


# .ravel()
#ravel  retuns view chnaging value Does affect original array
a13_reshaped_flat=a13_reshaped.ravel()
print(a13_reshaped_flat)
a13_reshaped_flat[1]=100 # changed in flatten array
print(a13_reshaped_flat)
print(a13) # also reflected in original array
```

```python
# operatiions on elements

# np.insert(original_array,index,axis)
# axis=none inserted in flatted version of array
# axis=0 inserted row wise
#axis=1 inserted colomun wise
print(a13)
new_a13=np.insert(a13,1,102,axis=None) # retuns another array with updated value no change in original array
print(f"original array: {a13}")
print(f"another updated array: {new_a13}")
print(len(a13))

# 2d array
print(f"original array: {a13_reshaped}")
arr1=np.insert(a13_reshaped,1,[102,110],axis=0) # row wise
print(f"another updated array(row wise):\n {arr1}")
arr1=np.insert(a13_reshaped,1,[102,110,129],axis=1) # col wise
print(f"another updated array(col wise):\n {arr1}")



```

```python
 #np.append(original_array,index,axis)
# axis=none appended in flatted version of array
# axis=0 appended row wise
#axis=1 appended colomun wise
#Append karte waqt arrays ki dimensions same aur shape compatible honi chahiye.


print(a13)
new_a13=np.append(a13,102,axis=None) # retuns another array with updated value no change in original array
print(f"original array: {a13}")
print(f"another updated array: {new_a13}")

# 2d array
print(f"original array: {a13_reshaped}")
arr1=np.append(a13_reshaped,[[102,110]],axis=0) # row wise
print(f"another updated array(row wise):\n {arr1}")
arr1=np.append(a13_reshaped,[[102],
                             [110],
                              [129]],axis=1) # col wise
print(f"another updated array(col wise):\n {arr1}")
```

```python
# concatenate
a1=np.array([1,2,3])
a2=np.array([4,5,6])
a3=np.concatenate((a1,a2))
print(a3)
a3=np.concatenate((a1,a2),axis=0)
print(a3)
a1=np.array([[1,2],[3,4]])
a2=np.array([[4,5],[6,7]])
a3=np.concatenate((a1,a2),axis=1)
print(a3)



```

```python
# stacking
# ''' vstack((arr1,arr2)) row wise
# hstack() col wise'''

a1=np.array([1,2,3,4])
a2=np.array([4,6,8,20])

print(np.vstack((a1,a2))) # vertically
print(np.hstack((a1,a2))) # horizontally

# 2d array
a1=np.array([[1,2],[3,4]])
a2=np.array([[4,6],[8,20]])
print(np.vstack((a1,a2))) # vertically
print(np.hstack((a1,a2))) # horizontally
```

```python
# splitting
"""
equal splitting
np.split(arr,split_size)
 np.hsplit() # col wise
 np.vsplit()
"""

a1=np.array([1,2,3,4,4,6,8,20])
print(np.split(a1,2))

a1=np.array([[1,2],[3,4],[4,6],[8,20]])
print(a1)
print(np.vsplit(a1,2)) # vertically
print(np.hsplit(a1,2))
```
